```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ConfiguraÃ§Ã£o do estilo
plt.style.use('default')
sns.set_palette("Set2")

# FunÃ§Ã£o auxiliar para calcular taxa de churn (%)
def churn_rate(df, col):
    return (
        df.groupby(col)["Churn"]
        .value_counts(normalize=True)
        .mul(100)
        .rename("ChurnRate")
        .reset_index()
        .query("Churn == 'Yes'")
    )

# FunÃ§Ã£o auxiliar para calcular taxa de churn completa (todas as categorias)
def churn_rate_complete(df, col):
    churn_data = df.groupby(col)["Churn"].apply(lambda x: (x == 'Yes').mean() * 100).reset_index()
    churn_data.columns = [col, 'ChurnRate']
    return churn_data

# FunÃ§Ã£o para adicionar valores nos grÃ¡ficos de barra
def add_value_labels(ax, bars, values, format_str='{:.1f}%'):
    """Adiciona rÃ³tulos de valores nas barras"""
    for bar, value in zip(bars, values):
        if ax.get_xlabel():  # Barras verticais
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                   format_str.format(value), ha='center', va='bottom',
                   fontweight='bold', fontsize=10)
        else:  # Barras horizontais
            width = bar.get_width()
            ax.text(width + width*0.01, bar.get_y() + bar.get_height()/2.,
                   format_str.format(value), ha='left', va='center',
                   fontweight='bold', fontsize=10)

# 1. Boxplot de mensalidade x churn
plt.figure(figsize=(8,6))
box_plot = sns.boxplot(data=df_completo, x="Churn", y="Charges.Monthly",
                      palette=['lightblue', 'lightcoral'])
plt.title("DistribuiÃ§Ã£o da Mensalidade por Churn", fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Mensalidade (R$)", fontsize=12)

# Adicionar estatÃ­sticas no boxplot
medians = df_completo.groupby('Churn')['Charges.Monthly'].median()
for i, (category, median) in enumerate(medians.items()):
    plt.text(i, median, f'Mediana: R${median:.0f}',
             horizontalalignment='center', verticalalignment='bottom',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8),
             fontweight='bold', fontsize=10)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 2. Churn por tipo de internet
plt.figure(figsize=(8,6))
internet_data = churn_rate(df_completo, "InternetService")
bars = plt.bar(internet_data["InternetService"], internet_data["ChurnRate"],
               color=['#FF9999', '#66B2FF', '#99FF99'], alpha=0.8, edgecolor='black', linewidth=1)

# Adicionar valores nas barras
add_value_labels(plt.gca(), bars, internet_data["ChurnRate"])

# Adicionar linha da mÃ©dia
avg_churn = df_completo['Churn'].value_counts(normalize=True)['Yes'] * 100
plt.axhline(y=avg_churn, color='red', linestyle='--', alpha=0.7, linewidth=2)
plt.text(0.5, avg_churn + 2, f'MÃ©dia Geral: {avg_churn:.1f}%',
         fontweight='bold', color='red', ha='center')

plt.title("Taxa de Churn por Tipo de Internet", fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Tipo de ServiÃ§o de Internet", fontsize=12)
plt.ylabel("Taxa de Churn (%)", fontsize=12)
plt.ylim(0, max(internet_data["ChurnRate"]) * 1.15)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 3. Churn por mÃ©todo de pagamento
plt.figure(figsize=(10,6))
payment_data = churn_rate(df_completo, "PaymentMethod")
bars = plt.barh(payment_data["PaymentMethod"], payment_data["ChurnRate"],
                color=['#FFB366', '#66FFB2', '#B366FF', '#FF66B2'],
                alpha=0.8, edgecolor='black', linewidth=1)

# Adicionar valores nas barras
add_value_labels(plt.gca(), bars, payment_data["ChurnRate"])

# Adicionar linha da mÃ©dia
avg_churn = df_completo['Churn'].value_counts(normalize=True)['Yes'] * 100
plt.axvline(x=avg_churn, color='red', linestyle='--', alpha=0.7, linewidth=2)
plt.text(avg_churn + 2, len(payment_data)/2, f'MÃ©dia\nGeral:\n{avg_churn:.1f}%',
         fontweight='bold', color='red', va='center', ha='left')

plt.title("Taxa de Churn por MÃ©todo de Pagamento", fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Taxa de Churn (%)", fontsize=12)
plt.ylabel("MÃ©todo de Pagamento", fontsize=12)
plt.xlim(0, max(payment_data["ChurnRate"]) * 1.2)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Preparar dados absolutos (contagem)
senior_counts = df_completo.groupby(["SeniorCitizen", "Churn"]).size().unstack(fill_value=0)

# Mapear 0/1 para labels
senior_counts.index = senior_counts.index.map({0: "NÃ£o Idoso", 1: "Idoso"})

# Criar grÃ¡fico empilhado
ax = senior_counts.plot(
    kind="bar",
    stacked=True,
    figsize=(8,6),
    color=["#99CCFF", "#FF9999"],  # azul = nÃ£o cancelou, vermelho = cancelou
    edgecolor="black",
    linewidth=1
)

# Adicionar valores absolutos dentro das barras
for container in ax.containers:
    ax.bar_label(container, label_type='center', fontsize=10, color='white', fontweight='bold')

# Adicionar porcentagens no topo de cada barra
totals = senior_counts.sum(axis=1)
for i, total in enumerate(totals):
    churn_pct = senior_counts.loc[senior_counts.index[i], "Yes"] / total * 100
    ax.text(i, total + total*0.02,
            f"Churn = {churn_pct:.1f}%",
            ha='center', va='bottom', fontsize=10, fontweight='bold')

# Ajustar tÃ­tulo e labels
plt.title("Churn por Faixa EtÃ¡ria (Volume Absoluto e ProporÃ§Ã£o)", fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Categoria", fontsize=12)
plt.ylabel("NÃºmero de Clientes", fontsize=12)
plt.legend(title="Churn", labels=["NÃ£o Cancelou", "Cancelou"])
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 5. Churn por tipo de contrato
plt.figure(figsize=(8,6))
contract_data = churn_rate(df_completo, "Contract")
bars = plt.bar(contract_data["Contract"], contract_data["ChurnRate"],
               color=['#FF6B6B', '#4ECDC4', '#45B7D1'], alpha=0.8,
               edgecolor='black', linewidth=1)

# Adicionar valores no topo das barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,  # posiÃ§Ã£o um pouco acima da barra
             f'{height:.1f}%', ha='center', va='bottom',
             fontweight='bold', fontsize=10)

# Adicionar linha da mÃ©dia
avg_churn = df_completo['Churn'].value_counts(normalize=True)['Yes'] * 100
plt.axhline(y=avg_churn, color='red', linestyle='--', alpha=0.7, linewidth=2)
plt.text(1, avg_churn + 3, f'MÃ©dia Geral: {avg_churn:.1f}%',
         fontweight='bold', color='red', ha='center')

# Adicionar contagem de clientes
counts = df_completo['Contract'].value_counts()
for bar, contract in zip(bars, contract_data["Contract"]):
    count = counts[contract]
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height()/2,
             f'n = {count}', ha='center', va='center',
             fontweight='bold', fontsize=10, color='white',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='black', alpha=0.7))

plt.title("Taxa de Churn por Tipo de Contrato", fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Tipo de Contrato", fontsize=12)
plt.ylabel("Taxa de Churn (%)", fontsize=12)
plt.ylim(0, max(contract_data["ChurnRate"]) * 1.2)
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#ðŸ“„Relatorio Final

A partir dos dados analisados, Ã© possÃ­vel chegar aos seguintes insights:
-Clientes que pagam mensalmente tem uma taxa de churn maior (42,7% destes deixaram de ser clientes).
-Clientes idosos, apesar de serem uma pequena parcela, tambÃ©m tem uma taxa de churn maior (41,7%).
-Clientes com uma mensalidade maior tendem a ter uma taxa de churn maior (os quartis 3 e 4 do boxplot de valor de mensalidade somam mais de 60% de taxa de churn).
-Clientes que optam por pagar com cheque eletrÃ´nico apresentam uma taxa de churn mais elevada se comparada aos outros mÃ©todos de Pagamento.
-Clientes que tem fibra Ã³tica como serviÃ§o tem uma taxa de churn mais significativa em comparaÃ§Ã£o aos que nÃ£o tem.

Com todos esses insights, Ã© possÃ­vel chegar a conclusÃ£o que a Empresa conta com um problema de fidelizaÃ§Ã£o de clientes, uma vez que as taxas mais elevadas de churn se encontram nos clientes mais novos, que pagam mensalmente, e com maiores mensalidades, indicando uma insatisfaÃ§Ã£o com os serviÃ§os prestados, e falta de benefÃ­cios para serem clientes a longo prazo, levando a uma elevada taxa de churn.
